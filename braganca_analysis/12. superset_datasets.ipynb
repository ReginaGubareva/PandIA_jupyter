```python
import pandas as pd
```

```python
df = pd.read_csv('data/dataset.csv', low_memory = False).fillna(0)
df['Installation_number'] = df['Installation_number'].astype(int)
df["Date"] = df["Year"].astype(str) + "/" + df["Month"].astype(str)
df["Date"] = pd.to_datetime(df["Date"], format='%Y/%m/%d')
df.head(5)
```

```python
df = df.drop(columns=['Year', 'Month', 'Consumer_number', 'Installation_zone', 'Installation_number'])
df.head(5)
```

```python
label_consumer = pd.read_csv('data/label_consumer.csv', low_memory = False)
label_consumer.head(5)
```

```python
df_2 = df.groupby(['Consumer_type', 'Date'], as_index=False).sum()
df_2.head(5)
```

```python
df_3 = df_2.merge(label_consumer, left_on='Consumer_type', right_on='Consumer_type')
df_3.head()
```

```python
df_3 = df_3.drop(columns=['Consumer_type'])
df_3 = df_3.groupby(['Date', 'Label'], as_index = False).sum()
len(df_3)
```

```python
df_3.to_csv('data/superset/clusters.csv', index=False)
```
